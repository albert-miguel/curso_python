```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=servico)
driver.get('file:///home/albert/project/python/curso_py/08_automacao_web/Pagina%20Hashtag.html')
```

```python
lista_elementos = driver.find_elements(By.CLASS_NAME, 'nav-link')
for ele in lista_elementos:
  if 'contato' in ele.text.lower():
    ele.click()
    break
```
