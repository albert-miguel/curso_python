```python
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
import numpy as np
```

#### Outros tipos de Gráficos:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
print(vendas)
```

```python
plt.plot(meses, vendas)
plt.axis([0, 50, 1000, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('vendas')
plt.show()
```

- Editando o gráfico de linha

```python
plt.plot(meses, vendas, 'ro')
plt.axis([0, 50, 900, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('vendas')
plt.show()
```

```python
plt.plot(meses, vendas, color='pink')
plt.axis([0, 50, 900, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('vendas')
plt.show()
```

```python
plt.plot(meses, vendas, 'r--')
plt.axis([0, 50, 900, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('vendas')
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Barras

```python
plt.bar(meses, vendas)
plt.show()
```

```python
plt.figure(1, figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.plot(meses, vendas, 'r--')
plt.subplot(1, 3, 2)
plt.plot(meses, vendas, color='pink')
plt.subplot(1, 3, 3)
plt.bar(meses, vendas)
plt.show()
```
